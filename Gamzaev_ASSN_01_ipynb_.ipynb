## Question #3
Given a list/vector of values, write code to count the instances of each unique value and store them in a data structure (hint… a dictionary will be helpful).
For example, if a list has 3 instances of 10.... the data structure should show {10: 3}
 - Python Input Data: [1,1,10,4,5,4,8,1,1,9,10,3, 3, 4, 1, 10, 11]


````python
#This code uses for-loop and if-statement to count the number of each unique value in the given list.
# The code works with any combination of values
mylist = [1,1,10,4,5,4,8,1,1,9,10,3, 3, 4, 1, 10, 11] # This is the list of unique numbers which we need to count
counter = {} # Creating the empty dictionary where we will put our unique values as keys and number of these values
for i in mylist: # Using for-loop here to check every value as "i" in our list and perform the next action
  if i in counter: # Using if-statement to check if current value "i" is already in our dictionary
    counter[i] +=1 # If we already have this value in our vocabulary then we just increase the number of it by 1
  else:
    counter[i] = 1 #If it is the first time we meet this value then we write just 1
print(counter) # Using function "print" to print our dictionary and check the number of each value in our list
````

## Question #7
Given the below input data types in Python, convert the following hash (dictionary) into a list that all have the same denomination. e.g. if a number is "1" in thousands, turn it into 1,000.
- Python: {10: "thousands", 1000: "ones", .0005: "millions", .000000324234: "trillions"}



Example Output:
```
[10000, 1000, 500, 324234]
```


````python
# Dictionary with numbers and their units
dict2 = {10: "thousands", 1000: "ones", .0005: "millions", .000000324234: "trillions"}

newlist=[] # new empty list to store the converted values

# Use converter to conver each unit (to turn everything into "ones")
converter = {"thousands": 1000, "ones": 1,"millions": 1000000, "trillions":1000000000000}
# Go through each key-value pair in the dictionary
for i,k in dict2.items():
  m = i* converter[k]
  m = int(m)
  newlist.append(m)

# Check our result as output using function "print"
print(newlist)

````

## Question #10
Taking in a list of multiple data types, sort the data in to a dictionary by the data type. The dictionary should consist of keys that correspond to a data type (e.g. int, float, str etc., the type/typeof functions should be useful), the values should be a list of all of the values of that data type
- Python: ["hey", 10, True, None, 23.43, 10.2, "second", 11,12, False, None]

````python
#Sorting values by data type

data = ["hey", 10, True, None, 23.43, 10.2, "second", 11,12, False, None]

# Create an empty dictionary to store results by type
# List for integer, float, string and etc values.
result = {"int" : [],
          "float" : [],
          "str" : [],
          "etc": []}

# Loop through each item in the original list
for i in data:
  if type(i) == int:
    result["int"].append(i) # Add integers to "int" list 
  else:
    if type(i) == float:
      result["float"].append(i) # The same here with float
    else:
      if type(i) == str:
        result["str"].append(i)  #The same here with sting
      else:
        result["etc"].append(i) #The same here with other
#Check the final result
result
````
