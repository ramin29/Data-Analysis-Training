```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data preparation

```python
# We need to read all our csv files and transform to the dataframe using Pandas
ci_09 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2009.csv')
ci_10 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2010.csv')
ci_11 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2011.csv')
ci_12 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2012.csv')
ci_13 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2013.csv')
ci_14 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2014.csv')
ci_15 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2015.csv')
ci_16 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2016.csv')
ci_17 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2017.csv')
ci_18 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2018.csv')
ci_19 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2019.csv')
ci_20 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2020.csv')
ci_21 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2021.csv')
ci_22 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2022.csv')
ci_23 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2023.csv')
ci_24 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2024.csv')
ci_25 = pd.read_csv('Class_05/Assignment/Data/Crime_Incidents_in_2025.csv')
```

```python
#Checking one of our dataframes to check whether it downloaded successfully
ci_25.head(5)
```

```python
# Now we need to merge all our dataframes from 2009 to 2025 in a single dataframe so we can work with full dataset
df = pd.concat([ci_09, ci_10, ci_11, ci_12, ci_13, ci_14, ci_15, ci_16, ci_17, ci_18,ci_19, ci_20, ci_21, ci_22, ci_23, ci_24, ci_25])
df.head(5)
```

```python
# Saving the REPORT_DAT column safely as datetime
df['REPORT_DAT'] = pd.to_datetime(df['REPORT_DAT'], utc=True, errors='coerce')
# Dropping rows with invalid/missing dates
df = df.dropna(subset=['REPORT_DAT']) 
# It is also useful to divide our date columns
# Extracting year from the report date
df['year'] = df['REPORT_DAT'].dt.year
# Createing a month column for easier monthly aggregation
df['month'] = df['REPORT_DAT'].dt.to_period('M').dt.to_timestamp() 
# Extracting the date (without time) for analyzing more common period
df['date'] = df['REPORT_DAT'].dt.date
# Extracting the hour of the day
df['hour'] = df['REPORT_DAT'].dt.hour
# Extracting the day of the week (Monday..Sunday)
df['weekday'] = df['REPORT_DAT'].dt.day_name()
df
```

# Part 1

## Line chart

```python
# Grouping the data by month and count the number of records for each month
monthly = df.groupby('month').size().rename('count')
# Setting up the figure size
plt.figure(figsize=(12,4))

# Ploting the monthly counts as a line chart
# marker='o' adds dots at each data point, linewidth=1 keeps the line thin
plt.plot(monthly.index, monthly.values, marker='o', linewidth=1)

# Adding title and axis labels
plt.title('Monthly crime counts from 2009 to 2025')
plt.xlabel('Month')
plt.ylabel('Number of incidents')

# Adjusting layout so everything fits good
plt.tight_layout()
plt.show()
```

## Insight 1:

Based on the gotten line chart, we can see that the highest number of crimes in DC occurred between 2014 and 2016. Moreover, there is a clear downward trend when analyzing the entire period.

# Part 2

## Horizontal bar chart

To indentify the 10 most common crime types we plotted horizontal bar chart

```python
# Finding the 10 most common offence types
top10 = df['OFFENSE'].value_counts().nlargest(10)
# Creating a horizontal bar plot of the top 10 offences
plt.figure(figsize=(10,6))

sns.barplot(x=top10.values, y=top10.index)
# Adding title and axis labels
plt.title('Top 10 OFFENSE (at all)')
plt.xlabel('Count')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

## Area chart

We use are chart to determine the most frequent types of crimes during the whole given period.

```python
top3 = df['OFFENSE'].value_counts().nlargest(3).index.tolist()
df_top3 = df[df['OFFENSE'].isin(top3)]
ts = df_top3.groupby(['month','OFFENSE']).size().unstack(fill_value=0)

plt.figure(figsize=(12,5))
ts.plot.area(alpha=0.6) 
plt.title('Trends of top 3 offences')
plt.xlabel('Month')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## Insight 2:

According to our bar chart and area chart we determined that MOTOR VEHICLE THEFT, THEFT F/AUTO and the THEFT/OTHER were the most frequesnt crimes during the whole period from 2009 to 2025.
