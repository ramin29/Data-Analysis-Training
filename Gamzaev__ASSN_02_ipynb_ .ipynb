## Question 1

Create a function that takes a list as input. The function should return TRUE if the list contains at least 1 even number, and false otherwise. The function should be assigned to name `contains_even`. Once you create the function use the function to determine if the following list contains any even numbers - `[11,13,15,16,19]`

## Explanation for question 1

Here we use a simple function where we set up for-loop and if-statement to check whether our numbers from the list are even numbers. It also add 'TRUE' and 'FALSE' text which we use in the final result.

````python
def contains_even(mylist):
  tr = 'TRUE'
  fl = 'FALSE'
  # Set up a for-loop to process each value in the list to perform the sequent steps
  for i in mylist:
    # Set-up an if-statement to check whether our numbers for the list are even numbers
    if i % 2 == 0:
      # Return TRUE if yes
      return tr
  # Return FALSE if no
  return fl
````

````python
#--- Dont change this!
print(contains_even([11,13,15,16,19]))
print(contains_even([11,13,15,19]))
````

## Question 2

Create a function named `counter` that given a list and an integer, will return how many times the integer appears inside the vector. The integer input should default to 10.

## Explanation for question 2

For solving this question we use a function where we set 10 as a default integer which we rewrite to the new one in the end. Then we set up a counnter to count how many times the integer appears inside the vector and use for-loop and if-statement to process it.

````python
def counter(input=10, vector_s=None):
  # Set up a counnter to count how many times the integer appears inside the vector
  number = 0
  # Set up a for-loop to process each value in the list to perform the sequent steps
  for i in vector_s:
    # Initialize an if-statement to check whether our current value from the list equals to input
    if i == input:
      # If our input equals to current value from the list then plus one to our counter
      number +=1
  return number
````

````python
#--- Dont change this!
print(counter(11, [11,13,15,16,11] ))
````

## Question 3

Without using any built-in functions, write a function named `max_finder` that takes a list as input and returns the largest value

## Explanation for Question 3



This code uses a function where is checked every value in the the given input list and returns the largest value

````python
# Initialize a function 'max_finder' and use it on the list 'cool_list'
def max_finder(cool_list):
  # Create a variable to the current largest possible value for later comparison with numbers from the list
  max_value = 0
  # Set up a for-loop to process each value in the list to perform the sequent steps
  for i in cool_list:
    # For each element in the list, use an if statement to compare it with current largest value in "max_value"
    if i > max_value:
      max_value=i
  # Return the final result as the largest value
  return max_value
````

````python
#--- Dont change this!
print(max_finder( [10,11,12,13,2] ))
````

## Question 4

Without using any built-in functions, write a function named `reverse_reverse` that takes a list as an input, and returns the same list, in reverse order

## Explanation for Question 4
For solving this task we use a  function which we named "reverse_reverse" to reverse the giving list

````python
# Use our standard function where "reverse_reverse" is  a name of the function and "mylist4" is a list wchich we will get later
def reverse_reverse(mylist4):
  # Create an empty list where we put our final reversed list
  result=[]
  # Create a counter for right indexing
  count4=0
  # Here we use for-loop to take every number in our list
  for i in mylist4:
    # When we take every number from the list we add 1 to the counter to follow the sequence
    count4+=1
    # Then we add  reversed number to our new list "result" using built-in function - append
    result.append(mylist4[-count4])
  # Then we close our function and finish it showing gotten list "result" using - return
  return result
````

````python
#--- Dont change this!
print(reverse_reverse( [1,2,3,4,5] ))
````

## Question 5

5. Write a function named `palindromer` that takes a atomic vector of characters as an input, and returns TRUE only if it is a Palindrome, and FALSE otherwise
e.g. c("H", "A", "N", "N", "A", "H") should be TRUE

e.g. c("B", "E", "N") should be FALSE

## Explanation for Question 5
Here we use almost the same function we used to solve Question 4 and add  if-statement to check wether the word is palindrome or not.

````python
# The function we used in the previous question
def palindromer(final_list):
  # Add variables with text 'TRUE' and 'FALSE' which we gonna use for the final result
  tr = 'TRUE'
  fl = 'FALSE'
  reverselist2=[]
  count5 = 0
  for i in final_list:
    count5 += 1
    reverselist2.append(final_list[-count5])
  # Create a simple if statement to check wheather our new list "reverselist2" equals to "final_list", where one is a reversed version of the other
  # If input list equals to the new list, it satisfies the condition of palindrome words
  if reverselist2 == final_list:
    return tr
  else: return fl
````

````python
#--- Dont change this!
print( palindromer( ["H", "A", "N", "N", "A", "H"] ))
print( palindromer( ["B", "E", "N"] ))
````
