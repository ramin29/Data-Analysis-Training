# 3. How many records exist under each of the following questions? Write code to find the answers to each question below **seperately**.

- 1a. Had an arrival delay of two or more hours
- 1b. Flew to Houston (IAH or HOU)

## **In this task I set a new variable for each sub-item**

```python
#-- a
# Sort by the column arrival delay from the lowest to the highest value
df_3a = df.sort_values('arr_delay', ascending = True)

# Create a new column where we transformed minutes to hours in column arrival delay
df_3a['arr_delay_hours'] = df_3a['arr_delay'] /  60

# Filter only by records (flights) having arrival delay more than 2 hours
df_3a = df_3a[df_3a['arr_delay_hours']>=2.0]

print(f'There are',len(df_3a), 'flights (records) having an arrival delay of two or more hours' )
```

## **1b. Flew to Houston (IAH or HOU)**

```python
#-- b

# Create 2 new dataframes for every airport in Houston and got information from column 'dest' in the main dataframe
df_3b_1 = df[df['dest'] == 'IAH']
df_3b_2 = df[df['dest'] == 'HOU']

# Create a new dataframe where we sum  the results and got one common dataframe for the whole flights to Houston
df_3b = df_3b_1 + df_3b_2
print(f'There are',len(df_3b), 'flights (records) to Houston' )
```

## **1c. Were operated by United, American, or Delta**

```python
# Use a specific function to find unique value in the column 'carrier' which contain keys with every airline
df['carrier'].unique()
```

```python
#-- c

# Again creeate 3 different dataframes for each airline and then sum the results in the one common dataframe to get the results for United, American and Delta airlines
df_3c_1 = df[df['carrier'] == 'UA']
df_3c_2 = df[df['carrier'] == 'AA']
df_3c_3 = df[df['carrier'] == 'DL']
df_3c = df_3c_1 + df_3c_2 + df_3c_3

# Print the result
print(f'There are',len(df_3c), 'flights (records) operated by United, American, or Delta Airlines' )
```

# **1d. Departed in summer (July, August, and September)**

```python
#-- d
# Create a new dataframe only with months before October (10)
df_3d = df[df['month'] < 10]

# Reset a new dataframe adding new condition by taking only month after June(6)
df_3d = df_3d[df_3d['month'] > 6]
# Now we have dataframe with July(7), August(8), and September(9)
print(f'Number of flights (records) departed in summer (July, August, and September):', len(df_3d) )
```

# **1f. Made up at least 60 minutes in flight time (hint.. you need to create a new column)**

```python
#-- f

# Create a new dataframe where calculate the difference between departure delay and arrival delay to understand how much time pilot made up
df_3f = df
df_3f['made_up_time'] = df_3f['dep_delay'] - df_3f['arr_delay']

# Filter by flights made up at least 60 minutes
df_3f = df_3f[df_3f['made_up_time'] >= 60]

print (f'Numbers of flights (records) where made up at least 60 minutes:' ,len(df_3f))
```

# **1g. Departed between midnight and 6am (inclusive)**

```python
#-- g
#Filter by departure time and consider two conditions in the same time
df_3g = df
# We start to filter according the first criteria
df_3g = df_3g[df_3g['dep_time'] <= 600]
# Filter according to the second criteria
df_3g = df_3g[df_3g['dep_time'] > 0]
print(f'Number of flights departed between midnight and 6am (inclusive):', len(df_3g))
```

# 5. What is the average departure delay time, for each airline? Who has the lowest? The highest? (hint.. group_by functions should be useful)

```python
#Group by airlines and find the average delay time for each
df_5 = df.groupby('carrier')['dep_delay'].mean()

print(df_5)

#Looking for airline with the lowest value
df_5_min = df_5.sort_values()
df_5_min_val = df_5_min.index[0]
print (f'The lowest departure delay time has', df_5_min_val, 'airline with the value of:', df_5_min.iloc[0])

#Looking for airline with the highest value using
df_5_max = df_5.sort_values(ascending = False)
df_5_max_val = df_5_max.index[0]
print (f'The highest departure delay time has', df_5_max_val, 'airline with the value of:', df_5_max.iloc[0])
```

**For qustion 6 I used logical operations for filtering the dataframe.**

**The link to the source that helped with the question 6 (Part - Filtering Data): https://realpython.com/pandas-dataframe/**

```python
#1 Here we created a new dataframe and filtered it using logical operations (the link to the source that helped with this question is in markdown above)

df_6 = df[(df['carrier'] == 'AA') |  (df['carrier'] == 'DL') | (df['carrier'] == 'UA')]
df_6
```

```python
#2

# Used function groupby and took into account two criteria for grouping such as year and carrier.
# Then using function "count" summed number of rows in column flight for American, Delta and United airlines
df_7 = df_6.groupby(['year', 'carrier'])['flight'].count()
df_7 = df_7.reset_index()
df_7
```

```python
#3

# In this task we got wide format using "pivot" function, where used columns from the previous task
# Fot widen the dataset we used 'year' as an index, 'carrier' as columns for a new dataframe and 'flight' as values for this dataframe
df_8 = df_7.pivot(index= 'year', columns = 'carrier', values = 'flight')
# Here we fix our dataframe and spplit column 'year' from indexing
df_8 = df_8.reset_index()
df_8
```

## 9. If you wanted to add the sector into stock_data, what would be the appropriate join? How many records would be in the result?

```python
import pandas as pd
```

```python
# Stock prices across 4 time periods
stock_data = {
    "ticker": ["AAPL", "AAPL", "AAPL", "AAPL",
               "MSFT", "MSFT", "MSFT", "MSFT",
               "GOOG", "GOOG", "GOOG", "GOOG",
               "TSLA", "TSLA", "TSLA", "TSLA",
               "AMZN", "AMZN", "AMZN", "AMZN"],
    "date": ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04"] * 5,
    "price": [172.0, 174.5, 175.5, 176.0,   # AAPL
              305.0, 308.2, 310.2, 312.0,   # MSFT
              130.5, 131.2, 132.8, 133.0,   # GOOG
              700.0, 710.0, 715.4, 720.0,   # TSLA
              3400.0, 3425.5, 3450.1, 3470.0]  # AMZN
}
stock_df = pd.DataFrame(stock_data)
stock_df.head(3)
```

```python
# Company info (single row per ticker)
company_data = {
    "ticker": ["AAPL", "MSFT", "GOOG", "TSLA", "NFLX", "JPM", "XOM", "DIS"],
    "sector": [
        "Technology",    # AAPL
        "Technology",    # MSFT
        "Technology",    # GOOG
        "Automotive",    # TSLA
        "Entertainment", # NFLX
        "Financials",    # JPMorgan
        "Energy",        # ExxonMobil
        "Entertainment"  # Disney
    ]
}
company_df = pd.DataFrame(company_data)
company_df
```

In this task we used joins to merge two dataframes.

```python
# Use left join to keep the information from dataframe stock_df and add only column 'sector' from company_df by common column 'ticker'
stock_merged = pd.merge(stock_df, company_df, how="left", on="ticker")

# Print the result
print(f'Number of records we got in this dataframe', len(stock_merged))

# Get our dataframe
print (stock_merged)
```
